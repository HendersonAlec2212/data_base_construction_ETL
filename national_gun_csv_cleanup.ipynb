```python
compare_first_last = new_nat_gunlaws.groupby(['state'])
first = compare_first_last.first().copy()
last = compare_first_last.last()
last.head(20)
```

```python
combined_first_last = pd.merge(first, last, how='inner', on=['state'], suffixes=("_start","_end"))
combined_first_last.head(20)
```

```python
combined_first_last['change_law_count'] =combined_first_last['lawtotal_end'] - combined_first_last['lawtotal_start']
combined_first_last.head()
```

```python
new_nat_gunlaws.to_csv('resources\\national_gun_laws_per_year\\clean_gun_laws_per_year.csv', index=True)
combined_first_last.to_csv('resources\\national_gun_laws_per_year\\summary_law_count_change.csv', index=True)
```
