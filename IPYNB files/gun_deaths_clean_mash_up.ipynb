```python
# estabilish list to hold newly iterated lists
years_list = []
for a in range(0,21):
    #redundant variable exchange
    i = a
    # list = i starting point & each nth value
    current_list = death_values[i::21]
    length = len(current_list)
    # append new, refined list to list
    years_list.append(current_list)
    
    # print data about current list

    print(current_list)
    print('-------')
    print(f'The length of the list is {length}')
    print(f'end of iteration, year {1999 + i}')
    print('-------')
    print('============')
```
